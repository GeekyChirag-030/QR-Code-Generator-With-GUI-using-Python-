```python
from tkinter import *
import tkinter as tk

from pyqrcode import*
from png import*

from tkinter.filedialog import*


from datetime import*
from tkinter.messagebox import*

from PIL import ImageTk, Image
from PIL.ImageTk import *
from PIL.Image import*

import os
```

```python
def URL():
    global S
    global Url
    S=0
    
    W=Tk()
    
    W['bg']='wheat'
    
    W.minsize(200,300)
    W.maxsize(1550,1550)
    
    Url=Data.get()
    Label(W,text="QR Code is Generated, To Save the File Please Click Here Below!",highlightcolor="grey",bd=6,bg="yellow",fg="black",font=("Times New Roman",22,"bold"),highlightthickness=5).pack()
    
    tk.Button(W,text="Submit",command=Submit2,font=('Arial',25,'bold'),bg='black',fg="wheat",relief="sunken",activebackground="blue",bd=7,activeforeground="yellow",highlightthickness=5,highlightcolor="grey").pack(padx=12,pady=12)
    
    Label(W,text="To Check That Your File is Save or Not Please Click Below , But please click here after click on the above Button ",highlightcolor="grey",bd=6,bg="yellow",fg="black",font=("Times New Roman",22,"bold"),highlightthickness=5).pack()
    
    tk.Button(W,text="Check",command=Check,font=('Arial',25,'bold'),bg='black',fg="wheat",relief="sunken",activebackground="blue",bd=7,activeforeground="yellow",highlightthickness=7,highlightcolor="grey").pack(padx=12,pady=12)
    
    
    W.mainloop()
    
    
def Submit2(): #This Basically Saves File in Computer
    global Path
    global S
    
    QR=create(Url) #QR Code is Created with global Variable Url
    S=asksaveasfilename(initialfile = 'C://Users//HP',defaultextension=".png",filetypes=[("All Files","*.*"),("Text Documents","*.txt")])
    New=""
    
    for I in S: #Logic For Generate the Right Path to Save image
        if I=="/":
            New=New+"//"
        else:
            New=New+I
    
    QR.png(New,scale=7)
    Path=New
   
    S=1#Shows if This Runs Then Only S=1
    
def Check(): #This Gives You the Successfull or Error Message
    
    if S==1:
        showinfo(title="Successful Message",message="File is Saved Successfully") #for Showing the messagebox which is kind of inforamtion box
            
        Ans=askquestion('Ask', 'Would You Like To Display Your QR?')

        if Ans=='yes': #Showing QR Code
            Label(Window,text="Your Generated QR Code is !",highlightcolor="grey",fg="black",bg="papayawhip",bd=4,font=("Times New Roman",25,"bold"),highlightthickness=3.7).pack(padx=7)

            F=Frame(Window).pack()
            img = ImageTk.PhotoImage(open(Path,"r"))

            label = Label(F,image=img) #Show Image After Setting it in Frame , image which Open in img 
            label.pack()

            Window.mainloop()
    else:
        showerror(title="Error",message="File is Not Saved")
        

def Quit(): #This Quits Your Application on The Basis of Answer
    
    Ss=askyesno(title="Quit",message="Do you want to quit this application ...? ")
    
    if(Ss==True):
        
        showinfo(title="Confirmation Message",message="  üëç  Ok , Application will quit !")
        Window.destroy()
        
def Delete():
    Ans=askyesno(title="Delete Request",message="Are you want to delete Your recently generated QR....?")
    
    if Ans==True:
        
        os.remove(Path)
        
        showinfo(title="Delete Request",message="Your file is deleted successfully...!")
        
    else:
        None

#Code from Where the Code Execution Starts         
        
Window=Tk()

Window.title("   *-*-*-*-*-*-*-*-*- QR Code Generator *-*-*-*-*-*-*-*-*-   ")
Window['bg']="#EEEEC5"

Window.iconbitmap('myqr.ico')

Window.geometry('450x550+350+450')
Window.minsize(200,300)
Window.maxsize(1550,1550)


L1=tk.Label(Window,text=" QR Code Generator ",font=('Times New Roman',45,"bold"),bd=12,highlightbackground='orange',bg='#FFFF82',highlightthickness=10,relief='sunken',cursor="dot").pack(side='top',pady=7
,padx=7)

Time=datetime.now()
Date=date.today()
Time=Time.strftime("%H:%M:%S")

L2=Label(Window,text=("Date & Time:- ",Date, Time),fg="#787878",bg="linen",font=('Cosmic Sans MS',35,'bold'),relief="groove",bd=5,highlightthickness=5,highlightbackground='#B2B222').pack(side='top',padx=12,pady=30)

L3=Label(Window,text=" *     Enter the URL (Uniform Resource Locator) Whose QR You Want to be Generated    *  ",font=('Arial',19,'bold'),bg='#EEEE76').pack(side='top',padx=12,pady=12)

Data=StringVar()

Entry(Window,cursor="hand2",width=40,font=('Times New Roman',25,'italic'),bd=4,textvariable=Data, selectbackground="darkblue",selectforeground="yellow",selectborderwidth=5).pack(padx=20,pady=20)

Button(Window,text="Submit",cursor="hand2",font=("Arial",30,"bold"),fg="linen",bg="black",activeforeground="blue",command=URL,activebackground="yellow",highlightthickness=3.5,highlightcolor="salmon",relief="groove",bd=4).pack(padx=20,pady=20)


Button(Window,text="Quit",cursor="hand2",font=("Arial",12,"bold"),fg="darkgreen",bg="red",activeforeground="blue",command=Quit,activebackground="yellow",highlightthickness=3.5,highlightcolor="salmon",relief="groove",bd=4).pack(side='left',padx=3,pady=3)

Button(Window,text="Delete",cursor="hand2",font=("Times New Roman",12,"bold"),bg="red",fg="green",activebackground="yellow",activeforeground="blue",command=Delete,bd=4,highlightthickness=3.5,highlightcolor="darkgrey",relief="groove").pack(side="right")
Window.mainloop()
```
